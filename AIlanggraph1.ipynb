```python
# Install required libraries (for Jupyter/Colab)
!pip install langgraph langchain python-dotenv transformers torch langchain-community

# Python standard imports
from typing import TypedDict, List, Literal

# LangChain core message import (note: plural 'messages')
from langchain_core.messages import HumanMessage

# LangGraph imports for building state graphs
from langgraph.graph import StateGraph, START, END

# Load environment variables from a .env file if available
from dotenv import load_dotenv
load_dotenv()

# Hugging Face model pipeline setup
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
```

```python


# Load a small, free text model (no API key needed)
hf_generator = pipeline("text2text-generation", model="google/flan-t5-large", max_length=100)

# Create an LLM instance compatible with LangChain
llm = HuggingFacePipeline(pipeline=hf_generator)

# Example state definition
class AgentState(TypedDict):
    messages: List[HumanMessage]
```

```python
def process(state: AgentState) -> AgentState:
    input_text = state["messages"][0].content
    prompt = f"You are a helpful assistant. Answer the question clearly:\n{input_text}"
    response = llm.invoke(prompt)
    print(f"\nAI: {response}")
    return state

graph = StateGraph(AgentState)
graph.add_node("process",process)
graph.add_edge(START, "process")
graph.add_edge("process", END)

app=graph.compile()
```

```python
user_input=input("Enter: ")
app.invoke({"messages": [HumanMessage(content=user_input)]})
```
